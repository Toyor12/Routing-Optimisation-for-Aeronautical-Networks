```python
# Import useful packages

!pip install utm
import pandas as pd
import numpy as np
import utm
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def read_flight_data(filename):
    ''' Read airplanes dataset as DataFrame'''
    airplanes_df = pd.read_csv(filename)
    return airplanes_df

def foot2km(feets):
    ''' Convert Altitude for feet to km'''
    km = 0.0003048 * feets
    return km

def distance(p1, p2):
    ''' Calculate the disctance between p1 and p2 in cartesian coordinate '''
    return np.linalg.norm(p2-p1)

def convert2cartesian(lat, lon, alt):
    ''' Convert Latitude(degree), Longitude(degree), Altitude(feet) to x(km),y(km),z(km) cartesian coordinates '''
    R = 6731                                     # Earth radius(km)
    lat, lon = np.deg2rad(lat), np.deg2rad(lon)  # Convert data from angle to radian
    alt = foot2km(alt)                           # Convert altitude from foot to km
    Rnew = R + alt                               # Distance of airplane to center of earth (new radius)
    # Cartesian coordinates
    x = Rnew * np.cos(lat) * np.cos(lon)
    y = Rnew * np.cos(lat) * np.sin(lon)
    z = Rnew * np.sin(lat)
    return pd.DataFrame({'x': x, 'y': y, 'z': z})
def read_rates_df():
    ''' Create a DataFrame table consist of data of Rates modes '''
    modes = [0, 1, 2, 3, 4, 5, 6, 7]
    colors = ['Black','Red','Orange','Yellow','Green','Blue','Pink','Purple']
    Thresholds = [740, 500, 400, 300, 190, 90, 35, 5.56]
    Rates = [0, 31.895, 43.505, 52.857, 63.970, 77.071, 93.854, 119.130]
    return pd.DataFrame({'Mode':modes, 'Color': colors, 'Switching': Thresholds, 'Rates': Rates})

def read_stations_df():
    names = ['LHR','EWR']
    latitude = [51.4700, 40.6895]
    longitude = [-0.4543, -74.1745]
    altitude = [81.73,8.72]

    stations_df = pd.DataFrame({'Station': names, 'Latitude': latitude, 'Longitude': longitude, 'Altitude': altitude})
    return stations_df
```

## 4.3.1 Single-Objective Optimization Problem
---------------------------------------------
The single-objective optimization problem aims to find a connection path between airplanes and ground stations that maximizes the transmission rate. To do this, we introduce the following parameters: $d_{ij}$, the distance between node $i$ and node $j$ (where node $i$ can be an airplane or ground station), and $r_{ij}$, the transmission rate between node $i$ and node $j$.

We define the optimization problem as follows: $x_{ij} \in {0,1}$, where $x_{ij} = 1$ indicates a connection between node $i$ and node $j$, and $x_{ij} = 0$ indicates there is no connection between the two nodes. $R_i$ is a variable that shows the final transmission rate between node $i$ and one of the ground stations. Specifically, we have $R_i = \min(r_{k_ng_n})$ for $n = 0,..., N_i$, where $k_0 = 0$, $k_n = g_{n-1}$, and $g_{N_i} \in GS$. This means that the final transmission rate for airplane $i$ is the minimum of the transmission rates of links starting at airplane $i$ and ending at one of the ground stations. Therefore, the optimization problem is:

$$ \max_{x_{ij}} \sum_{i \in A} R_i $$

To solve this optimization problem efficiently, we can use search algorithms such as Dijkstra's algorithm to find the shortest path. However, since the original Dijkstra's algorithm is designed to minimize the path, we need to modify it in two ways: first, we need to maximize the transmission rate instead of minimizing the path, and second, we need to update the new rate using the minimum rate of the path instead of the algebraic sum.

To calculate the distances between nodes, we need to put all the airplanes and ground stations in a single set of nodes, and then calculate the distance between each pair of nodes, saving the results in a distance matrix. The element $d_{ij}$ in this matrix represents the distance between node $i$ and node $j$. To calculate the transmission rates, we can define a function that takes the distance matrix as input and returns the corresponding matrix of transmission rates using a table of transmission rates. The element $r_{ij}$ in this matrix represents the transmission rate between node $i$ and node $j$.

```python
def calculate_distance(data1, data2 = []):
    ''' this function calculate the distance between all points of data1 and data2'''
    if len(data2) == 0:
        data2 = data1
    n1 = len(data1)
    n2 = len(data2)
    distances = np.zeros((n1, n2))
    for n in range(n1):
        for m in range(n2):
            distances[n,m] = distance(data1[n], data2[m])
    return distances

def calculate_transmission_rates(distances, rates_df):
    ''' This function get the distance and calculate the rate of transmission of data between points'''
    Thresholds = rates_df['Switching']
    Rates = rates_df['Rates']
    rates = np.zeros_like(distances)
    for n in range(len(Thresholds)):
        if n == 0:
            rates += Rates[n] * (distances > Thresholds[n])
        else:
            rates += Rates[n] * ((distances <= Thresholds[n-1]) & (distances > Thresholds[n]))
    return rates
```

```python
import networkx as nx

class Graph():
    def __init__(self, airplane_df, stations_df, solution):
        self.nodes_number = len(airplanes_df) + len(stations_df)
        self.airplanes_df = airplanes_df
        self.stations_df = stations_df
        self.solution = solution
        self.positions = np.vstack((airplanes_df[['x','y']].values.astype('float64'), stations_df[['x','y']].values.astype('float64')))
    def get_links(self):
        links = np.zeros((self.nodes_number, self.nodes_number))
        set_of_nodes = [[] for j in range(len(self.stations_df)+1)]
        for n in range(len(self.airplanes_df)):
            path = self.solution[n]['path']
            station = self.solution[n]['station']

            if self.solution[n]['transmission_rate'] > 0:
                set_of_nodes[station].append(n)
                for j in range(1, len(path)):
                    fr = path[j-1]
                    to = path[j]
                    links[fr,to] = 1
            else:
                set_of_nodes[len(stations_df)].append(n)

        # Seperate the links
        set_of_links = [[] for i in range(len(self.stations_df))]
        for i in range(self.nodes_number):
            for j in range(self.nodes_number):
                if links[i,j] == 1:
                    for s in range(len(stations_df)):
                        if i in set_of_nodes[s]:
                             set_of_links[s].append((i,j))
        return set_of_links
    def get_nodes(self):
        set_of_airplanes = [[] for j in range(len(self.stations_df)+1)]
        for n in range(len(self.airplanes_df)):
            station = self.solution[n]['station']
            if self.solution[n]['transmission_rate'] > 0:
                set_of_airplanes[station].append(n)
            else:
                set_of_airplanes[len(stations_df)].append(n)
        return set_of_airplanes
    def show_network(self, show_node_number):
        
        self.links = self.get_links()
        self.nodes = self.get_nodes()
        colors = ["tab:red", "tab:blue", "tab:orange", "tab:green","tab:purple", "tab:olive"]
        G = nx.Graph()
        pos = {}
        for node in range(self.nodes_number):
            position = (self.positions[node, 0], self.positions[node, 1])
            G.add_node(node, pos = position)
            pos[node] = position
        plt.figure(figsize=(15,10))
        for index in range(len(self.nodes)):
            nodes_set = self.nodes[index]
            nx.draw_networkx_nodes(G, pos = pos, nodelist = nodes_set, node_color = colors[index], node_size=20)
        for station in range(len(self.stations_df)):
            node = station + len(airplanes_df)
            nx.draw_networkx_nodes(G, pos = pos, nodelist = [node], node_color = colors[len(self.nodes)], node_size=150)
        for index in range(len(self.links)):
            links_set = self.links[index]
            nx.draw_networkx_edges(G,pos,edgelist = links_set, width = 1, alpha = 0.5, edge_color= colors[index])
        
        if show_node_number:
            airplanes_no = len(self.airplanes_df)
            labels = {airplanes_no + station: self.stations_df['Station'][station] for station in range(len(self.stations_df))}
            nx.draw_networkx_labels(G, pos, labels, font_size=10)
        plt.show()
```

## 4.3.3 Solver for one station
------------------------------------
If the target station is known, we can use a searching algorithm to find the best path with the maximum rate. The Routing Network class takes in the matrix of rates and the target station as input and calculates the optimal connection with the maximum rate between all the airplanes and the target station. Once the problem is solved, the class returns the $R_i$ values and the corresponding path for each airplane.



```python
class RoutingNetwork():
    '''
    This class is written to solve the problem using a
    searching algorithm to find the best way with maximum
    rate which can connect the airpalnes to selected station.
    We do this algorithm for all the stations, then check which
    one of stations has higher rate of end-to-end data rate transmission 
    and select that path for each airplane'''
    def __init__(self, matrix_rates):
        ''' Initialize the class with rates'''
        self.matrix_rates = matrix_rates
        self.max_rate = np.max(matrix_rates)
        size = matrix_rates.shape
        if size[0] != size[1]:
            print('Error, Graph must be square...')
        else:
            self.vertices = size[0]
        
    def maxRate(self, rates, selected):
        # Initialize maximum rate for next node
        max_ = -1
        # Search for the vertex with maximum rate, which is not selected before
        for vertex in range(self.vertices):
            if rates[vertex] > max_ and selected[vertex] == False:
                max_ = rates[vertex]
                max_index = vertex
        return max_index
        
    def dijkstra(self, station):
        '''
        This function has a similar algorithm as dijkstra algorithm,
        only to find the best path check the
        ** min(rates[vertex], matrix_rates[max_index][vertex]) **
        which is the algorithm for calculating the transmission of each path
        '''
        # Initialize the rates as 0
        rates = [0] * self.vertices
        # Set the rate for station as the maximum value of rates
        rates[station] = self.max_rate
        # print(rates)
        # Initialize the connection(path) between vertices
        path = [[i, station] for i in range(self.vertices)]
        
        # Initilize the selected vectors
        selected = [False] * self.vertices

        # Run the searching path to maximize the rate
        for iteration in range(self.vertices):
            # Pick the maximum rate vertex from
            # the set of vertices not yet processed.
            # max_index is always equal to station in first iteration
            max_index = self.maxRate(rates, selected)

            # Put the maximum rate vertex in the in the selected vertices
            selected[max_index] = True

            # Update dist value of the adjacent vertices
            # of the picked vertex only if the current
            # rate is less than new rate and
            # the vertex in not in selected vectors
            for vertex in range(self.vertices):
                if (self.matrix_rates[max_index][vertex] > 0 and selected[vertex] == False and rates[vertex] < min(rates[max_index], 
                    self.matrix_rates[max_index][vertex])):  
                    rates[vertex] = min(rates[max_index] , self.matrix_rates[max_index][vertex])
                    path[vertex] = [vertex, max_index]
        return rates, path
```

```python
def printResults(airplanes_df, stations_df, rates, finalSolution):
    ''' This file print the final solution with desired format as in description file explained'''
    airplanes_number = len(finalSolution)
    for airplane in range(airplanes_number):
        print("{}'{}', 'routing-path':".format('{',airplanes_df['Flight No.'][airplane]), end = '')
        path = finalSolution[airplane]['path']
        rate = finalSolution[airplane]['transmission_rate']
        station = finalSolution[airplane]['station']
        for node in range(1, len(path)):
            fr = path[node-1]
            to = path[node]
            if to >= airplanes_number:
                print("({},{}),".format(stations_df['Station'][station], rates[fr][to]), end = '')
            else:
                print("({},{}),".format(airplanes_df['Flight No.'][to], rates[fr][to]), end = '')
        print(' end-to-end data rate: {} Mbps {}'.format(rate,'}'))  
        print('*****************************************************')

class RoutingAeronauticalNetworks():
    ''' 
    This class is the main class that manage the solution of problem 1,
    In this class we get all the raw data as input, and calcualte the distances,
    transmission rates, and solve the problem for each station using RoutingNetwork
    class and finally find the best path between airplanes and stations,
    Also print the results in the format that is explained with desription file.
    '''
    def __init__(self, airplanes_df, stations_df, rates_df):
        ''' Initialize the class data'''
        self.airplanes_df = airplanes_df
        self.stations_df = stations_df
        self.stations_number = len(stations_df)
        self.airplanes_number = len(airplanes_df)
        self.rates_df = rates_df
        self.airplanes_pos = self.airplanes_df[['x','y','z']].values.astype('float64')
        self.stations_pos = self.stations_df[['x','y','z']].values.astype('float64')
        self.positions = np.vstack((self.airplanes_pos, self.stations_pos))
        self.distances = calculate_distance(self.positions)
        self.rates = calculate_transmission_rates(self.distances, self.rates_df)
        

    def getSolution(self, show):
        '''
        This function run the solver for each station and collect the results in RATES and PATHS,
        Finally it found the best solution in RATES and PATH for each airplane and save the results
        in finalSolution list. If show == True, then the function print the results
        '''
        RATES = []
        PATHS = []
        # Run RoutingNetworkSolver for each station
        for station in range(self.stations_number):
            source = self.airplanes_number + station        # Station id in the vertices
            rates, path = self.RoutingNetworkSolver(source) # Run the RoutingNetworkSolver for source
            RATES.append(rates)                             # Collect rates in RATES
            PATHS.append(path)                              # Collect path in PATHS
        
        # Retrieve the best solution for each airplane
        finalSolution = []
        for airplane in range(self.airplanes_number):
            opt_results = {}
            # Search for best station for current ariplane
            max_rate_station = 0                         
            max_rate = RATES[max_rate_station][airplane]
            for station in range(self.stations_number):
                if RATES[station][airplane] > max_rate:
                    max_rate = RATES[station][airplane]
                    max_rate_station = station
            opt_results['station'] = max_rate_station                             # best station for airplane
            opt_results['transmission_rate'] = RATES[max_rate_station][airplane]  # best rate for airplane
            opt_results['path'] = PATHS[max_rate_station][airplane]               # best path to connect the airplane to station
            finalSolution.append(opt_results)                                     # Collect the results in finalSolution
        # Print results in show == True
        if show:
            printResults(self.airplanes_df, self.stations_df, self.rates, finalSolution)
        return finalSolution
        
    def RoutingNetworkSolver(self, station):
        # Create object from RoutingNetwork
        routing_model = RoutingNetwork(self.rates)
        rates, path = routing_model.dijkstra(station)
        
        # Change the path in routing model to standard format (list of nodes that are in the path)
        PATH = [[i] for i in range(self.airplanes_number)]
        for airplane in range(self.airplanes_number):
            fr = airplane
            to = path[fr][1]
            PATH[airplane].append(to)
            while to != station:
                fr = to
                to = path[fr][1]
                PATH[airplane].append(to)
        return rates, PATH
    
```

# 4.4 Multi-objective optimization problem
------------------------------------------

As said in above section, the single-objective optimization can lead to suboptimal results. In this study, we aim to address this issue by considering a multi-objective optimization problem for routing optimization in aeronautical networks. Specifically, we aim to maximize the end-to-end data transmission rate and minimize the end-to-end latency for each airplane. To this end, we assume that each link in the network has a fixed delay of 50 ms. Thus, if an airplane is connected to a station through 10 links, the total delay for this connection would be 500 ms.

To solve this problem, we define a matrix of delays, with the element (i,j) representing the delay between node i and node j. It should be noted that the delay is 50 ms for all nodes, except for the self-loop (i,i).


Also we need to define a rule, which select the best path. In order to select the optimal path, the following rules are applied:

a) If $d_j >= d_i$ and $t_j < t_i$, then path $p_j$ is selected as the optimal path.

b) If $d_j > d_i$ and $t_j <= t_i$, then path $p_j$ is selected as the optimal path.

c) If $d_j > 0$ and $t_j < t_i$, then path $p_j$ is selected as the optimal path.

d) If none of the above conditions are met, the last optimal path is retained.

As previously mentioned, the Routing Network algorithm is designed to find the optimal path for a single target station. However, in this problem, there are two stations, thus we need to modify the algorithm to consider multiple target stations. To do this, we must determine the best path and rate for each target station for each airplane. For example, if the best route connecting airplane 0 to station 216 has a rate of 52.857 Mbps and the path connecting airplane 0 to station 217 has a rate of 77.071 Mbps, we would select the connection between airplane 0 and station 217 due to its higher transmission rate.

To implement this, we have defined a new class that runs the Routing Network algorithm for each station, collects the results, and then selects the optimal path for each airplane. The final solution consists of the "final rate" and the corresponding "path" for each airplane.

### Extended dijkstra algorithm
<pre> 1  <b>function</b> ExtendedDijkstra(<i>Graph</i>, <i>matrix_rates</i>, <i>station</i>):
 2      
 3      <b>for each</b> vertex <i>v</i> in <i>Graph.Vertices</i>:
 4          rate[<i>v</i>] ← 0
 5          path[<i>v</i>] ← v
 6          add <i>v </i> to <i>Q</i>
 7      dist[<i>station</i>] ← max(rates)
 8      
 9      <b>while</b> <i>Q</i> is not empty:
10          <i>u</i> ← vertex in <i>Q</i> with maximum rate[u]
11          remove u from <i>Q</i>
12          
13          <b>for each</b> neighbor <i>v</i> of <i>u</i> still in <i>Q</i>:
14              <i>new_rate</i> ← min(rate[<i>u</i>] + matrix_rates(<i>u</i>, <i>v</i>))
15              <b>if</b> <i>new_rate</i> &gt; rate[<i>v</i>]:
16                  rate[<i>v</i>] ← <i>new_rate</i>
17                  add <i> u </i> to path[<i>v</i>]
18
19      <b>return</b> rate[], path[]
</pre>


```python
def delay_of_link():
    return 50.0
def calculate_transmission_delays(distances):
    ''' This function get the distance and calculate the delay of transmission of data between points'''
    delays = delay_of_link() * np.ones_like(distances)
    for i in range(delays.shape[0]):
        delays[i,i] = 0
    return delays
class MultiObjectiveRoutingNetwork():
    '''
    This class is written to solve the problem using a
    searching algorithm to find the best way with maximum
    rate which can connect the airpalnes to selected station.
    We do this algorithm for all the stations, then check which
    one of stations has higher rate of end-to-end data rate transmission 
    and select that path for each airplane'''
    def __init__(self, matrix_rates, matrix_delays):
        ''' Initialize the class with rates'''
        self.matrix_rates = matrix_rates
        self.matrix_delays = matrix_delays
        self.max_rate = np.max(matrix_rates)
        
        size = matrix_rates.shape
        if size[0] != size[1]:
            print('Error, Graph must be square...')
        else:
            self.vertices = size[0]
        
    def maxRate(self, rates, selected):
        # Initialize maximum rate for next node
        max_ = -1
        # Search for the vertex with maximum rate, which is not selected before
        for vertex in range(self.vertices):
            if rates[vertex][0] > max_ and selected[vertex] == False:
                max_ = rates[vertex][0]
                max_index = vertex
        return max_index
        
    def multiobjective_searchpath(self, station):
        '''
        This function has a similar algorithm as dijkstra algorithm,
        only to find the best path check the
        ** min(rates[vertex], matrix_rates[max_index][vertex]) **
        which is the algorithm for calculating the transmission of each path
        '''
        # Initialize the rates as 0
        rates = [[0,1e3] for vertex in range(self.vertices)]
        # Set the rate for station as the maximum value of rates
        rates[station] = [self.max_rate, 0]
        # print(rates)
        # Initialize the connection(path) between vertices
        path = [[i, station] for i in range(self.vertices)]
        
        # Initilize the selected vectors
        selected = [False] * self.vertices

        # Run the searching path to maximize the rate
        for iteration in range(self.vertices):
            # Pick the maximum rate vertex from
            # the set of vertices not yet processed.
            # max_index is always equal to station in first iteration
            max_index = self.maxRate(rates, selected)

            # Put the maximum rate vertex in the in the selected vertices
            selected[max_index] = True

            # Update dist value of the adjacent vertices
            # of the picked vertex only if the current
            # rate is less than new rate and
            # the vertex in not in selected vectors
            for vertex in range(self.vertices):
                if (self.matrix_rates[max_index][vertex] > 0 and self.matrix_rates[max_index][vertex] > 0 and 
                    selected[vertex] == False and self.isPareto(rates, max_index, vertex)):
                    rates[vertex][0] = min(rates[max_index][0] , self.matrix_rates[max_index][vertex])
                    rates[vertex][1] = rates[max_index][1] + self.matrix_delays[max_index][vertex]
                    path[vertex] = [vertex, max_index]
        return rates, path
    def isPareto(self, rates, max_index, vertex):
        old_rate = rates[vertex][0]
        old_delay = rates[vertex][1]
        new_rate = min(rates[max_index][0] , self.matrix_rates[max_index][vertex])
        new_delay = rates[max_index][1] + self.matrix_delays[max_index][vertex]
        return paretoSelection(old_rate, new_rate, old_delay, new_delay)

def paretoSelection(old_rate, new_rate, old_delay, new_delay):
    if (old_rate < new_rate and old_delay >= new_delay) or (old_rate <= new_rate and old_delay > new_delay):
        return True
    elif (old_rate >= new_rate and old_delay <= new_delay):
        return False
    elif new_rate > 0 and (old_delay > new_delay):# or (new_rate-old_rate) >= 0.2*(old_delay-new_delay)):
        return True
    else:
        return False
```

```python
class MultiObjectiveRoutingAeronauticalNetworks():
    ''' 
    This class is the main class that manage the solution of problem 1,
    In this class we get all the raw data as input, and calcualte the distances,
    transmission rates, and solve the problem for each station using RoutingNetwork
    class and finally find the best path between airplanes and stations,
    Also print the results in the format that is explained with desription file.
    '''
    def __init__(self, airplanes_df, stations_df, rates_df):
        ''' Initialize the class data'''
        self.airplanes_df = airplanes_df
        self.stations_df = stations_df
        self.stations_number = len(stations_df)
        self.airplanes_number = len(airplanes_df)
        self.rates_df = rates_df
        self.airplanes_pos = self.airplanes_df[['x','y','z']].values.astype('float64')
        self.stations_pos = self.stations_df[['x','y','z']].values.astype('float64')
        self.positions = np.vstack((self.airplanes_pos, self.stations_pos))
        self.distances = calculate_distance(self.positions)
        self.rates = calculate_transmission_rates(self.distances, self.rates_df)
        self.delays = calculate_transmission_delays(self.distances)
        

    def getSolution(self, show):
        '''
        This function run the solver for each station and collect the results in RATES and PATHS,
        Finally it found the best solution in RATES and PATH for each airplane and save the results
        in finalSolution list. If show == True, then the function print the results
        '''
        RATES = []
        PATHS = []
        # Run RoutingNetworkSolver for each station
        for station in range(self.stations_number):
            source = self.airplanes_number + station        # Station id in the vertices
            rates, path = self.RoutingNetworkSolver(source) # Run the RoutingNetworkSolver for source
            RATES.append(rates)                             # Collect rates in RATES
            PATHS.append(path)                              # Collect path in PATHS
        
        # Retrieve the best solution for each airplane
        finalSolution = []
        for airplane in range(self.airplanes_number):
            opt_results = {}
            # Search for best station for current ariplane
            pareto_station = 0                         
            max_rate = RATES[pareto_station][airplane][0]
            min_delay = RATES[pareto_station][airplane][1]
            for station in range(self.stations_number):
                if paretoSelection(max_rate, RATES[station][airplane][0], min_delay, RATES[station][airplane][1]):
                    max_rate = RATES[station][airplane][0]
                    min_delay = RATES[station][airplane][1]
                    pareto_station = station
            opt_results['station'] = pareto_station                             # best station for airplane
            opt_results['transmission_rate'] = RATES[pareto_station][airplane][0]  # best rate for airplane
            opt_results['path'] = PATHS[pareto_station][airplane]               # best path to connect the airplane to station
            finalSolution.append(opt_results)                                     # Collect the results in finalSolution
        # Print results in show == True
        if show:
            printResults(self.airplanes_df, self.stations_df, self.rates, finalSolution)
        return finalSolution
        
    def RoutingNetworkSolver(self, station):
        # Create object from RoutingNetwork
        mo_routing_model = MultiObjectiveRoutingNetwork(self.rates, self.delays)
        rates, path = mo_routing_model.multiobjective_searchpath(station)
        
        # Change the path in routing model to standard format (list of nodes that are in the path)
        PATH = [[i] for i in range(self.airplanes_number)]
        for airplane in range(self.airplanes_number):
            fr = airplane
            to = path[fr][1]
            PATH[airplane].append(to)
            while to != station:
                fr = to
                to = path[fr][1]
                PATH[airplane].append(to)
        return rates, PATH

```

## Problem 1
--------------
The single-objective optimization approach to routing in aeronautical networks has several limitations. One issue is that the focus is solely on maximizing the end-to-end data transmission rate, regardless of the number of connections in the path or the distance between the airplane and the target station. This can result in airplanes being connected to stations that are far away but have many connections, as illustrated in the above figure.

Another limitation is that the connection rate is a discrete value, which means that there may be multiple paths with the same rate. However, these paths may differ in other features, such as distance or energy consumption, and only the first one found by the algorithm is chosen, while the others are ignored.

On the positive side, the solver used in this approach is fast and capable of finding the global optimal solution, not just a local one. Additionally, the method can be extended to multi-station problems, making it suitable for larger networks.

##Problem 2
-----------------
One weakness of the multi-objective optimization approach is that it does not consider the distance between nodes, and the delays are fixed, which means it cannot find the shortest path with a high transmission rate. Instead, it only minimizes the number of connections, rather than the length of the path.

Additionally, as with the single-objective optimization, the discrete values of the transmission rate and delays may result in multiple optimal solutions for the problem. This method cannot choose between these solutions based on other features that may be important, such as the length of the path.

On the positive side, the solver used for this optimization problem is fast and is able to find the global optimal solution, rather than getting stuck in a local minimum. The final solution also minimizes the number of connections and connects airplanes to the nearest station.

## 6.0 Conclusion
---------------------
In conclusion, the multi-objective optimization approach resulted in more reasonable solutions compared to the single-objective optimization method in this routing optimization for aeronautical networks. The extended Dijkstra algorithm proved to be a suitable solver for this routing problem due to its speed and efficiency. However, the use of fixed delays for each link as a criterion for optimization may not adequately consider factors such as the distance and energy required for routing in the network. To further optimize the routing network, it may be necessary to include an additional objective such as the length of the path. Alternatively, if the transmission rate and delay are prioritized over the length of the path, the objectives can be prioritized accordingly.
